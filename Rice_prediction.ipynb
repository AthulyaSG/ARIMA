# Non-stationary

Why make a non-stationary series stationary before forecasting?
Forecasting a stationary series is relatively easy and the forecasts are more reliable.

An important reason is, autoregressive forecasting models are essentially linear regression models that utilize the lag(s) of the series itself as predictors.

We know that linear regression works best if the predictors (X variables) are not correlated against each other. So, stationarizing the series solves this problem since it removes any persistent autocorrelation, thereby making the predictors(lags of the series) in the forecasting models nearly independent.

# Question 2

In one sentence, interpret the result of the AD-Fuller Test

# Answer

Coupling the AD-Fuller check together

```python
# For single country

# To check stationary or not
# Augmented Dickey Fuller test (ADF Test) is a common statistical test used to test
# whether a given Time series is stationary or not

from statsmodels.tsa.stattools import adfuller

# Load the time series data from a CSV file
data = pd.read_csv('C:\\Users\\dell\\AI Assignment\\ARG_4.csv')

# Perform the ADF test
result = adfuller(data['value'])

# Extract and print the test statistics and p-value
ADF_test = result[0]   #result[0] returns the test statistic value
p_value = result[1]    #result[1] corresponds to the p-value in the ADF test
crit_value = result[4]['5%'] #result[4] corresponds to the critical value [retrieve the critical value for a given level of confidence (e.g., 5%)]
print('ADF Test Statistic:', ADF_test)
print('p-value:', p_value)
print('Critical value:', crit_value)

# Interpret the test results
if p_value < 0.05 and ADF_test < crit_value :
    print('The time series is stationary')
else:
    print('The time series is non-stationary')
```

```python
# Grouped area and year together

from statsmodels.tsa.stattools import adfuller

# Load the time series data from a CSV file
data = pd.read_csv('C:\\Users\\dell\\AI Assignment\\final_2.csv')

# Group the data by location and year, and calculate the mean of the values for each group
grouped_data = data.groupby(['area', 'year'])['value'].mean()

# Loop over each location and run the ADF test on the time series
for location in grouped_data.index.levels[0]: #to loop over the unique values in the first level of a multi-level index
    location_data = grouped_data.loc[location]
    result = adfuller(location_data)
    print(f"Location {location}:")
    print(f"ADF Statistic: {result[0]}")   #result[0] returns the test statistic value
    print(f"p-value: {result[1]}")    #result[1] corresponds to the p-value in the ADF test
    print(f"Critical values: {result[4]}") #result[4] corresponds to the critical value [retrieve the critical value for a given level of confidence (e.g., 5%)]
    if result[1] < 0.05 and result[0] < result[4]['5%']:
        print('The time series is stationary')
    else:
        print('The time series is non-stationary')
```

```python
# Coupling AD-Fuller check with rolling statistics

def adfuller_test(timeseries, transformation_type, window = 3):
    movingAverage = timeseries.rolling(window).mean()
    movingSTD = timeseries.rolling(window).std()
    
    orig = plt.plot(timeseries, color='cornflowerblue', label='Original')
    mean = plt.plot(movingAverage, color='firebrick', label='Rolling Mean')
    std = plt.plot(movingSTD, color='limegreen', label='Rolling Std')
    plt.legend(loc = 'best')
    plt.title(f'{transformation_type} Rolling Statistics', size = 14)
    plt.show(block=False)
    
    adf = adfuller(timeseries, autolag='AIC') # autolag='AIC': to automatically determine the lag length for the ADF test
 
    print(f'ADF Statistic: {adf[0]:.3}')
    print(f'p-value: {adf[1]:.3}')
    print(f'#Lags Used: {adf[2]}')
    print(f'#Number of Observations Used: {adf[3]}')
    print("##################################")  
    print('Critical Values:')
    
    for key, ts in adf[4].items():
         print(f'{key}: {ts:.3}')
    print("##################################")
    
    if adf[0] > adf[4]["5%"]:
        print("ADF > Critical Values")
        print ("Failed to reject null hypothesis, time series is non-stationary.")
    else:
        print("ADF < Critical Values")
        print ("Reject null hypothesis, time series is stationary.")
        
adfuller_test(indexed_data, 'Non transformation')
```

# Data Transformation to Achieve Stationary

```python
# Logarithmic Transformation

indexed_data_log_scaled = np.log(indexed_data)

plt.plot(indexed_data_log_scaled, color = 'cornflowerblue')
plt.xlabel('Year', size = 12)
plt.ylabel('Rice Production', size = 12)
plt.title("After Logarithmic Transformation", size = 14)
```

```python
# Removing Trend in Moving Average

moving_avg = indexed_data_log_scaled.rolling(window=12).mean()
indexed_data_log_scaled_ma = indexed_data_log_scaled - moving_avg

indexed_data_log_scaled_ma.dropna(inplace=True) #Remove NAN values

plt.plot(indexed_data_log_scaled_ma, color = 'cornflowerblue')
plt.xlabel('Year', size = 12)
plt.ylabel('Rice Production', size = 12)
plt.title("After Moving Average", size = 14)
```

```python
# Exponential Decay Transformation
# Exponential weightes moving average
# halflife: number of time periods for the weighting to decrease by half

indexed_data_log_scaled_ma_ed = indexed_data_log_scaled_ma.ewm(halflife=12, min_periods=0, adjust=True).mean()
indexed_data_log_scaled_minus_ma_ed = indexed_data_log_scaled_ma - indexed_data_log_scaled_ma_ed

plt.plot(indexed_data_log_scaled_minus_ma_ed, color='cornflowerblue')
plt.xlabel('Year', size = 12)
plt.ylabel('Rice Production', size = 12)
plt.title("After Exponential Decay Transformation", size = 14)
```

```python
adfuller_test(indexed_data_log_scaled_minus_ma_ed, "After Exponential Decay Transformation", window = 12)
```

```python
# Time Shift Transformation
# scaled.shift(): to shift the index by a specified number of periods

indexed_data_log_scaled_shift = indexed_data_log_scaled - indexed_data_log_scaled.shift()

plt.plot(indexed_data_log_scaled_shift, color='cornflowerblue')
plt.xlabel('Year', size = 12)
plt.ylabel('Rice Production', size = 12)
plt.title("After Time Shift Transformation", size = 14)
```

```python
indexed_data_log_scaled_shift.dropna(inplace=True) #Remove NAN values
adfuller_test(indexed_data_log_scaled_shift, "After Time Shift Transformation", window = 12)
```

# AR (AutoRegression) Model

```python
ar_values = indexed_data_log_scaled_minus_ma_ed.values

#Split the data into two part such that our models can be used 
#to predict the last 10 days of rice production in the dataset given all prior observations
train = ar_values[1:len(ar_values)-10]
test = ar_values[len(ar_values)-10:]

#Build the AR model
model = ARIMA(train, order=(2,1,0))
AR_model = model.fit()

#Predict the rice production for the next 10 days 
predictions = AR_model.predict(start=len(train), end=len(train) + len(test)-1, dynamic=False)
ar_score = mean_squared_error(test, predictions)
print(f'AR MSE: {ar_score:.4}')

#When an AR model is fit to a time series, the coefficients of the model are estimated using a technique, MSE
```

```python
from statsmodels.tsa.arima_model import ARIMA
from sklearn.metrics import mean_squared_error

model = ARIMA(train, order=(0,1,2))
start_params = [0.1, 0.1, 0.1] #holds a list of starting values for the parameters of a statistical model
MA_model = model.fit(start_params=start_params)

predictions = MA_model.predict(start=len(train), end=len(train)+len(test)-1, dynamic=False)
ma_score = mean_squared_error(test, predictions)
print(f'MA MSE: {ma_score:.4}')
```
